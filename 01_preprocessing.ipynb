```python
import numpy as np
import pandas as pd
import sklearn
sklearn.__version__
```

```python
data = {'size':  ['XL', 'L', 'M', 'L', 'M'], 'color' : ["red", "green", "blue", "green", "red"],
        'gender' : ["female", "male", "male", "female", "female"], 'price' : [199.0, 89.0, 99.0, 129.0, 79.0],
        'weight' : [500, 450, 300, 380, 410], 'bought' : ['yes', 'no', 'yes', 'no', 'yes']}
df_raw = pd.DataFrame(data=data)
df_raw
```

```python
#kopia danych:
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')
df['weight'] = df['weight'].astype('float')
df.info()
```
