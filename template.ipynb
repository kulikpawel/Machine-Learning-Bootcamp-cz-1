```python
import pandas as pd
import numpy as np
size  = pd.array(['XL', 'L', 'M', 'L', 'M'])
color = pd.array(["red", "green", "blue", "green", "red"])
gender = pd.array(["female", "male", "male", "female", "female"])
price = pd.array([199.0, 89.0, 99.0, 129.0, 79.0])
weight = pd.array([500, 450, 300, 380, 410])
bought = pd.array(["yes", "no", "yes", "no", "yes"])
df = pd.DataFrame({'size':size, 'color':color, "gender": gender, 'price':price, 'weight':weight, 'bought':bought})
df
```

```python
df = pd.get_dummies(df)
```

```python
df
```

```python
print(df.columns)
df = df.drop(["size_L", "color_blue", "gender_female", "bought_no"], axis=1)
df
```

```python
srednia = df['price'].mean()
odchylenie = df['price'].std(ddof=0)
df['price'] = (df['price'] - srednia)/odchylenie
srednia = df['weight'].mean()
odchylenie = df['weight'].std(ddof=0)
df['weight'] = (df['weight'] - srednia)/odchylenie
df
```
